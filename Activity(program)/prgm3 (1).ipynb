```
def square(n):
    return n ** 2
print(square(5))
```

```
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

print("17 is prime:", is_prime(17))
print("4 is prime:", is_prime(4))
print("2 is prime:", is_prime(2))
print("1 is prime:", is_prime(1))
```
