PANDAS

```python
import pandas as pd
```

```python
groceries=pd.Series(data=[30,6,'yes','no'], index= ['eggs','apple','milk','bread'])
print(groceries)
```

```python
x='bananas'in groceries
print(x)
```

```python
x='milk'in groceries
print(x)
```

```python
print('How many egggs do we need to buy:',groceries['eggs'])
```

```python
print('How many eggs and apples do we need to buy:', groceries.loc[['eggs','apple']])
```

```python
print('Do we need milk and bread:', groceries[['milk','bread']])
```

```python
groceries['eggs']=2
print(groceries)
```

```python
groceries.drop('apple')
print(groceries)
```

DATA FRAME

```python
items={'Bob':pd.Series(data=[245,25,55], index=['bike','pants','watch']), 'Alice':pd.Series(data=[40,110,500,45], index=['book','glasses','bike','pants'])}
print(items)
```

```python
shopping_carts=pd.DataFrame(items)
print(shopping_carts)
```

```python
data={'integer':[1,2,3], 'float':[4.5,8.3,9.6]}
print(data)
df=pd.DataFrame(data)
print(df)
```

```python
items['Bob'].pop('watch')
print(items)
```
