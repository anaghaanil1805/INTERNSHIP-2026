```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test)=tf.keras.datasets.cifar100.load_data()
from tensorflow.keras.applications import VGG16

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
x_train,y_train=x_train[:100],y_train[:100]
x_test,y_test=x_test[:100],y_test[:100]
```

```python
x_train = tf.image.resize(x_train, (224, 224))
x_test = tf.image.resize(x_test, (224, 224))
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
y_train=to_categorical(y_train.reshape(-1), num_classes=100)
y_test=to_categorical(y_test.reshape(-1), num_classes=100)
```

```python
base_model=VGG16(weights='imagenet',include_top=False,input_shape=(224,224,3))
```

```python
for layer in base_model.layers:
  layer.trainable=False
```

```python
x=base_model.output
x=Flatten()(x)
x=Dense(512,activation='relu')(x)
x=Dropout(0.5)(x)
prediction=Dense(100, activation='softmax')(x)
```

```python
model= Model(inputs=base_model.input,outputs=prediction)
```

```python
model.compile(optimizer=Adam(learning_rate=1e-4),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
for layer in base_model.layers[-4:]:
  layer.trainable=True
```

```python
model.compile(optimizer=Adam(learning_rate=1e-5),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history_fine=model.fit(x_train,y_train,batch_size=32,epochs=10,validation_split=0.2)
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
print("Test accuracy", test_acc)
print("Test loss", test_loss)
```
