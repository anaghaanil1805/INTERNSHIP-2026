```python
def check_even_odd(num):
  if num % 2==0:
    return 'even'
  else:
    return 'odd'
print(check_even_odd(7))
print(check_even_odd(10))
```

```python
def count_vowels(text):
  vowels='aeiouAEIOU'
  count=0
  for char in text:
    if char in vowels:
      count+=1
  return count
print(count_vowels("Anagha"))
```

```python
def simple_model(score):
  if score>50:
    return 'pass'
  else:
    return 'fail'
print(simple_model(85))
```

```python
def checking_voiting_age(age):
  if age>18:
    return 'can vote'
  else:
    return 'cannot vote'
print(checking_voiting_age(20))
```

```python
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)
print(factorial(5))
```
