TWO SUM

```python
def two_sum(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        difference = target - num

        if difference in seen:
            return [seen[difference], i]

        seen[num] = i

    return []

nums = [2, 7, 11, 15]
target = 9

print(two_sum(nums, target))
```

*Explanation:*

*Store numbers in a dictionary.
Check whether the required pair already exists.
Time Complexity: O(n)*

*Goal: Find two numbers whose sum equals a target.*

Section 2: k-NN & Distance Logic

**1. K Closest Points to Origin**

```python
import math

def k_closest(points, k):
    points.sort(key=lambda p: math.sqrt(p[0]**2 + p[1]**2))
    return points[:k]

points = [[1, 3], [-2, 2], [5, 8]]
k = 2

print(k_closest(points, k))
```

*Goal: Find points closest to origin (0,0)*


How it works:
Use distance formula:


​

Sort points based on distance
Take first k


Trick:

We don’t actually need square root (but used for clarity)


**2. Find Nearest Point with Same X or Y**

```python
def nearest_valid_point(x, y, points):
    min_distance = float('inf')
    index = -1

    for i, (px, py) in enumerate(points):
        if px == x or py == y:
            distance = abs(px - x) + abs(py - y)

            if distance < min_distance:
                min_distance = distance
                index = i

    return index

print(nearest_valid_point(3, 4, [(1,2), (3,1), (2,4), (3,4)]))
```

**Explanation:**

Goal: Find closest point with same x OR y coordinate

How it works:
Loop through points


Check:
px == x OR py == y

If valid:
Compute Manhattan distance


|x1 - x2| + |y1 - y2|
Track minimum distance


Section 3: Clustering & Grouping
**1. Group Anagrams**

```python
from collections import defaultdict

def group_anagrams(words):
    groups = defaultdict(list)

    for word in words:
        key = ''.join(sorted(word))
        groups[key].append(word)

    return list(groups.values())

words = ["eat", "tea", "tan", "ate", "nat", "bat"]

print(group_anagrams(words))
```

**Explanation:**

Goal: Group words with same letters

How it works:
Sort each word → use as key
Store words in dictionary

Example:
"eat", "tea", "ate" → key = "aet"

Output:
[['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]

**2. Number of Islands**

```python
def num_islands(grid):
    if not grid:
        return 0

    rows = len(grid)
    cols = len(grid[0])
    count = 0

    def dfs(r, c):
        if r < 0 or c < 0 or r >= rows or c >= cols or grid[r][c] == "0":
            return

        grid[r][c] = "0"

        dfs(r+1, c)
        dfs(r-1, c)
        dfs(r, c+1)
        dfs(r, c-1)

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1":
                count += 1
                dfs(r, c)

    return count

grid = [
    ["1","1","0","0"],
    ["1","0","0","1"],
    ["0","0","1","1"]
]

print(num_islands(grid))
```

**Explanation:**

Goal: Count separate groups of '1's in grid

How it works:
Traverse grid

When you find '1':
Start DFS (Depth First Search)
Convert connected '1's → '0' (mark visited)

Key idea:

Each DFS = one island


Movement:
Up, Down, Left, Right

Section 4: Matrix & Image Processing

**1. Flood Fill**

```python
def flood_fill(image, sr, sc, color):
    original = image[sr][sc]

    if original == color:
        return image

    def dfs(r, c):
        if r < 0 or c < 0 or r >= len(image) or c >= len(image[0]):
            return

        if image[r][c] != original:
            return

        image[r][c] = color

        dfs(r+1, c)
        dfs(r-1, c)
        dfs(r, c+1)
        dfs(r, c-1)

    dfs(sr, sc)

    return image

image = [
    [1,1,1],
    [1,1,0],
    [1,0,1]
]

print(flood_fill(image, 1, 1, 2))
```

**Explanation:**

**Goal**: Replace connected pixels with new color

How it works:
Start from (sr, sc)
Save original color

Use DFS:
Replace all connected same-color cells

Example:

Like paint bucket tool in MS Paint

**Set matrix zeroes**

```python
def set_zeroes(matrix):
    rows = set()
    cols = set()

    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j] == 0:
                rows.add(i)
                cols.add(j)

    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if i in rows or j in cols:
                matrix[i][j] = 0

    return matrix

matrix = [
    [1,1,1],
    [1,0,1],
    [1,1,1]
]

print(set_zeroes(matrix))
```

***Explanation:***

Goal: If a cell is 0 → make its row & column 0

How it works:

First pass:
Store rows & columns containing 0

Second pass:
Set those rows & columns to 0

Why 2 steps?

Avoid affecting results while scanning

Section 5: Sliding Window
1. Maximum Average Subarray

```python
def find_max_average(nums, k):
    window_sum = sum(nums[:k])
    max_sum = window_sum

    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i-k]
        max_sum = max(max_sum, window_sum)

    return max_sum / k

nums = [1,12,-5,-6,50,3]
k = 4

print(find_max_average(nums, k))
```

**Explanation:**

Goal:

Find subarray of size k with max average

How it works:

Calculate first window sum

Slide window:

new_sum = old_sum + next_element - removed_element
Why efficient?

No recomputing sum each time → O(n)

2. Sliding Window Maximum

```python
from collections import deque

def max_sliding_window(nums, k):
    dq = deque()
    result = []

    for i in range(len(nums)):
        while dq and dq[0] <= i-k:
            dq.popleft()

        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()

        dq.append(i)

        if i >= k-1:
            result.append(nums[dq[0]])

    return result

print(max_sliding_window([1,3,-1,-3,5,3,6,7], 3))
```

**Explanation**

Goal:

Find max in every window of size k


How it works:

Use deque (double-ended queue)
Store indices of useful elements only

Logic:

Remove elements out of window
Remove smaller elements (they can't be max)
Front of deque = max

```python
students = [
    ["Harry", 37.21],
    ["Berry", 37.21],
    ["Tina", 37.2],
    ["Akriti", 41],
    ["Harsh", 39]
]

scores = sorted(set(score for name, score in students))

second_lowest = scores[1]

names = sorted(name for name, score in students if score == second_lowest)

for name in names:
    print(name)
```

**Explanaion:**

Sort them

Find second lowest

Print names with that score
